# **DETR**

```python
import torch
import yaml

from context_utils import PathContext
from pathlib import Path
```

```python
with PathContext('image_analyzer/'):
    from detr.inference import inference_on_video
    from detr.models.detr import DETR
    
    # constants
    INPUT_PATH = 'examples/catatonic_schizophrenic_interview.mp4'
    MODEL_WEIGHTS = 'demo'
    
    # path compensation because of PathContext
    INPUT_PATH = Path('../', INPUT_PATH)
    
    
    # load config
    with open('configs/flickr_faces.yaml', 'r') as f:
        config = yaml.safe_load(f)

    # load model
    print('Loading DETR...')
    model = DETR(config['dataset']['num_classes'],
                 config['model']['dim_model'],
                 config['model']['n_heads'],
                 n_queries=config['model']['n_queries'],
                 head_type=config['model']['head_type'])

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    
    # load weights
    if MODEL_WEIGHTS == 'demo':
        model.load_demo_state_dict('data/state_dicts/detr_demo.pth')
    else:
        state_dict = torch.load(MODEL_WEIGHTS, map_location=device)['state_dict']
        model.load_state_dict(state_dict)
    
    # inference on video
    output_path = (INPUT_PATH.parent / (INPUT_PATH.stem + '_out')).with_suffix('.mp4')
    inference_on_video(model, INPUT_PATH, output_path)
```

#### The result of the visual inference is a video, the demo video result can be seen in examples/catatonic_schizophrenic_interview_out.mp4
